```python
def burger():
    bur_bill = 0

    burger_choices={'varities_in_burgers':'Cost(Rs)','small_burger':30,'medium_burger':45,'larger_burger':60,'chef_special':70}
    bur_keys=list(burger_choices.keys())
    bur_val=list(burger_choices.values())
    for i in range(len(burger_choices)):
        print(i,bur_keys[i],bur_val[i])


    while True:
        item2 = input('select the burger number you would like to have for your meal:("type:00" to quit)')#use exception handling here for valueError
        if not(item2=='00'):
            bur_count = input('how many would you like to have:')
        else:
            break

        if item2=="00":
            break
        else:
            b = int(item2)
            bur_cost = bur_val[b]
            a = int(bur_cost) * int(bur_count)
            bur_bill = int(a) + int(bur_bill)

    return bur_bill
```
